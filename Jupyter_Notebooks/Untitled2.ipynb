```python
import os
import pandas as pd
lengths = {}
for entry in os.scandir('/home/MarkF/DivideConquer/Results/Cancer_Types_Norm'):
    if 'AllCancer' not in entry.path and 'GPL570_norm' not in entry.path:
        df = dd.read_csv(entry.path)
        lengths[entry.path.split('Type_')[1].replace('.csv', '')] = len(df)
    
```

```python
lengths['GPL570'] = 21592
plot_df = pd.DataFrame.from_dict(lengths, orient='index').reset_index()
plot_df.columns = ['cancer type', 'count']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.barplot(data=plot_df, x='cancer type', y='count')
plt.xticks(rotation=90)
plt.show()
```
